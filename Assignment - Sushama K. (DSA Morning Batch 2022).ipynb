# # Question no.5 - Perform meaningful visualizations using the dataset. Bring at least 3 visualizations.


1. BOXPLOT

```python
import seaborn as sns
sns.boxplot(iris_data)
```

2. HISTOGRAM

```python
plt.hist(iris_data['SL'],color="y",rwidth=0.9,cumulative=True)
plt.title("Distribution of SL",fontsize = 25)
plt.xlabel('SL',fontsize = 15)
plt.ylabel('Frequency',fontsize = 15)
plt.show()
```

```python
plt.hist(iris_data['SW'],color="r",rwidth=0.9,cumulative=True)
plt.title("Distribution of SW",fontsize = 25)
plt.xlabel('PL',fontsize = 15)
plt.ylabel('Frequency',fontsize = 15)
plt.show()
```

```python
plt.hist(iris_data['PL'],color="g",rwidth=0.9,cumulative=True)
plt.title("Distribution of PL",fontsize = 25)
plt.xlabel('PL',fontsize = 15)
plt.ylabel('Frequency',fontsize = 15)
plt.show()
```

```python
plt.hist(iris_data['PW'],color="b",rwidth=0.9,cumulative=True)
plt.title("Distribution of PW",fontsize=25)
plt.xlabel('PL',fontsize = 15)
plt.ylabel('Frequency',fontsize = 15)
plt.show()
```
